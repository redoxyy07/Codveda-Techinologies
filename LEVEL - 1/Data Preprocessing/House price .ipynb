### 10. Standardize the numerical features

#### StandardScaler
    - SandardScaler is used to standardize the numerical features to have a mean=0 & standard deviation=1.
    - It ensures that each features contribute equally.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

a_scaled = sc.fit_transform(a)
```
