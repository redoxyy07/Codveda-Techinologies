### 10. Encoding Categorical Features

#### Label Encoder
    - Label Encoder is used to convert the text data into numerical data.
    - It is useful because ML model can't directly work with the text data.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
a['Text'] = le.fit_transform(a['Text'])
```

```python
le1 = LabelEncoder()
a['Sentiment'] = le1.fit_transform(a['Sentiment'])
```

```python
le2 = LabelEncoder()
a['User'] = le2.fit_transform(a['User'])
```

```python
le3 = LabelEncoder()
a['Hashtags'] = le3.fit_transform(a['Hashtags'])
```

```python
le4 = LabelEncoder()
a['Platform'] = le4.fit_transform(a['Platform'])
```

```python
le5 = LabelEncoder()
a['Country'] = le5.fit_transform(a['Country'])
```

### 11. Standardize the features

#### StandardScaler
    - StandardScaler is used to standardize the numerical features to have a mean=0 & standard deviation=1.
    - It helps to prevent features with large ranges from dominating the features with small ranges.
    - It also ensures that all features contribute equally.

```python
a_enco = a[['Sl no','Text','Sentiment','Hashtags','User','Platform','Retweets','Likes','Country','Year','Month','Day','Hour']]
a_enco.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
a_scaled = sc.fit_transform(a_enco)
```

```python
a_final = pd.DataFrame(a_scaled, columns=a_enco.columns, index=a_enco.index)
```

```python
a_final.head()
```

### 12. Final Dataset

```python
a_final = pd.concat([a['Timestamp'],a_final],axis=1)
```
