### 8. Encoding Categorical Features

#### Label Encoder
    - Label Encoder is used to convert text data into numerical datas.
    - It is done because most of the ML models can't directly work with the text datas.
    - Therefore we need to encode them.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
a['State'] = le.fit_transform(a['State'])
```

```python
le1 = LabelEncoder()
a['International plan'] = le1.fit_transform(a['International plan'])
```

```python
le2 = LabelEncoder()
a['Voice mail plan'] = le2.fit_transform(a['Voice mail plan'])
```

```python
le3 = LabelEncoder()
a['Churn'] = le3.fit_transform(a['Churn'])
```

```python
a.head()
```

### 9. Standardize the numerical features

#### Standard Scaler
    - StandardScaler is used to standardize the numerical features to have a mean=0 & standard deviation=1.
    - It ensures that all features contribute equally.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
a_scaled = sc.fit_transform(a)
```

### 10. Final Dataset

```python
a_final = pd.DataFrame(a_scaled, columns=a.columns, index=a.index)
```
