### 5. Visualization

#### a. Bar Plot

```python
plt.figure(figsize=(8,6))
sns.barplot(x=a['Churn'], y=a['Total day minutes'])
plt.title("Bar Plot for Total Day Minutes by Churn status", fontweight='bold')
plt.xlabel("Churn Status", color='blue')
plt.ylabel("Total Day Minutes", color='blue')
plt.legend(['Total Churn'],title='Legend')
plt.savefig("Box plot for Total Day Minutes by Churn Status.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation :
    This Box plot is used to show the distribution of Total Dat Minutes by Churn Status.
    The maximum Churned customers has total around 220 minutes a day.
    And the maximum Unchurned customers has a total around 175 minutes a day
    From this graph it is cleared that Customers having maximum Total Day Minutes are likely to Churned.

#### b. Line Chart

```python
plt.figure(figsize=(8,4))
gd = a.groupby('Customer service calls')['Total day minutes'].mean()
plt.plot(gd.index, gd.values, marker='o', linestyle='-', color='red')
plt.title('Customer Service Calls vs Total Day Minutes', style='italic')
plt.xlabel("Customer Service Calls", fontsize=10)
plt.ylabel("Average Total Day Minutes", fontsize=10)
plt.savefig("Line chart for Customer Service Calls vs Total Day Minutes.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretration :
    The Line Plot shows the average Total Day Minutes vs Number of Customer Service Calls.
    It is found that Customers those have made 0-4 Customer Service Calls have medium Day minutes.
    And at the right side the graph increases suddenly which means that Customers that made more than 7 Service calls have the highest Total Day Minutes.
    The minimum number of Total Day Minutes has around 7 Customer Service Calls.

#### c. Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=a['Total intl minutes'], y=a['Total intl charge'], hue=a['Churn'])
plt.title("Scatter plot for Total International Call Minutes vs Charges", color='red')
plt.xlabel("Total International Minutes", fontsize=9)
plt.ylabel("Total International Charges", fontsize=9)
plt.legend(title='Churn Status')
plt.savefig("Scatter Plot for Total International Call Minutes vs Charges.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation :
    The Scatter Plot is used to show the relationship between Total International Minutes vs Total International Charges.
    The plot has two colors : Orange -> Churn Customers & Blue -> Unchurned Customers.
    The plot shows a strong linear relationhip means with increase in Total international minutes, Total International Charges also increases.
    But these features couldn't clearly helps in distinguishing Churned and Unchurned Customers.
