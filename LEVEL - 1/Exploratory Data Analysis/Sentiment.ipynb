### 5. Summary Statistics

```python
num_col = a.select_dtypes(include=['number'])
```

#### a) Mean :
    Mean is the sum of all values divided by the number of values.

```python
num_col.mean()
```

#### b) Median :
    Median is the middle value when the values are sorted.

```python
num_col.median()
```

#### c) Mode :
    Mode is the value that occurs most of the times in a dataset or a sequence.

```python
num_col.mode().iloc[0]
```

#### d) Standard Deviation :
    It measures the amount of variation from the mean.

```python
num_col.std()
```

### 6. Visualizating Data Distribution

#### a) Histogram

```python
plt.figure(figsize=(10,6))
a['Likes'].hist(bins=20, grid=False, rwidth=0.9, edgecolor='Black')
plt.title("Histogram of Likes")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.show()
```

#### Interpretation :
    This histogram is used to represent the distribution of Likes.
    Most of the posts are getting around 30-50 likes.
    Most of the likes are from the left side that indicates that the graph is 'Right Skewed'.
    The Maximum likes is about 80 and the minimum number of likes is 10.

#### b) Box Plot

```python
plt.figure(figsize=(10,4))
sns.boxplot(y=a[a['Retweets'] > 0]['Retweets'])
plt.title("Box plot for Retweets")
plt.ylabel("Number of Retweets")
plt.show()
```

#### Interpretation :
    This Box Plot shows the distribution of Retweets.
    The median values of Retweets is around 22.
    The InterQuartile Range is about 17 to 25 retweets.
    The minimum retweet is about 7 and maximum retweet is about 35.
    

#### c) Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=a['Likes'],y=a['Retweets'],hue=a['Platform'])
plt.title("Scatter Plot for Likes vs Retweets on Platforms")
plt.xlabel("Number of Likes")
plt.ylabel("Number of Retweets")
plt.show()
```

#### Interpretation :
    This Scatter plot shows the relationship between Number of Likes and Number of Retweets on different platforms.
    Here different colours represents different platforms :
        Blue & Red -> Twitter, Yellow -> Instagram, Green -> Facebook.
    This graph clearly indicates that there are more number of likes and retweets on Twitter than other platforms.

### 7. Correlation

```python
print(a.corr(numeric_only=True))
```

```python
cm = a.corr(numeric_only=True)
plt.figure(figsize=(12,6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='.2f')
plt.title("Correlation of numerical features")
plt.show()
```

#### Interpretation :
    The heatmap shows the correlation between numerical features.
    Darker shades = Highly Correlated, Lighter Shades = Lightly Correlated.
    As we can se Likes & Retweets are highly correlated which means if one increases the other one also increases.
