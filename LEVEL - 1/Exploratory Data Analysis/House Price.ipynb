### 6. Summary Statistics

```python
num_col = a.select_dtypes(include=['number'])
```

#### a) Mean :
    - Mean is the sum of values divided by the total number of values.

```python
print("Mean : ")
num_col.mean().round(2)
```

#### b) Median :
    Median is the middle most value in a sorted sequence of values.

```python
print("Median : ")
num_col.median().round(2)
```

#### c) Mode :
    Mode is the value that occurs the most frequent time in a sequence of values.

```python
print("Mode : ")
num_col.mode().iloc[0].round(2)
```

#### d) Standard Deviation :
    Standard Deviation is the amount of variation from the mean.

```python
num_col.std().round(2)
```

### 7. Visualizing Data Distribution

#### a) Histogram :

```python
plt.figure(figsize=(8,6))
sns.histplot(a['MEDV'], kde=True, edgecolor='Black')
plt.title("Histogram for House Price Distribution")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.show()
```

#### Interpretation :
    This Histogram shows the distribution of House Price.
    From this graph, it is observed that most house prices are around $18k - $25k.
    The Maximum price of house is about $50k and the minimum house price is about $5k.
    Most of the values are observed in the left side which indicates it is 'Right Skewed'.

#### b) Box Plot :

```python
plt.figure(figsize=(8,6))
sns.boxplot( y=a['CRIM'] )
plt.title("Box plot for Crime Rate")
plt.ylabel("Crime Rate ")
plt.show()
```

#### Interpretation :
    The Box Plot shows the distribution of Crime Rate across different towns.
    This box plot clearly indicates that there are many number of outliers present in the dataset.
    And most of the outliers are in the top means these are having most number of Crime Rate than others.

#### c) Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=a['MEDV'], y=a['CRIM'], hue=a['CHAS'])
plt.title("Scatter plot for House Price vs Crime Rate")
plt.xlabel("House Price")
plt.ylabel("Crime Rate")
plt.show()
```

#### Interpretation :
    The Scatter Plot shows the relation between 'MEDV'(House Price) & 'CRIM'(Crime Rate), with different colors showing whether the house is near to the Charles River.
    Here, Light color -> Far from Charles River, Darker color -> Nearer to Charles River.
    From this graph it is observed that lower house price are having hgher Crime Rate.
    And the houses with higher prices are having less Crime Rate.

### 8. Correlation :

```python
print(a.corr(numeric_only=True))
```

```python
cm = a.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='.2f')
plt.title("Correlation for numerical features")
plt.show()
```

#### Interpretation :
    The Heatmap is used to determine the correaltion among different features.
    From this graph it is found that RAD & NOX has 0.99 correlation, which means houses with high accessibility with the highways(RAD) have more Pollution(NOX).
