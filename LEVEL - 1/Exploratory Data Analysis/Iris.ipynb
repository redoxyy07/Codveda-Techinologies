### 6.  Visualizing Data Distribution

#### a)Histogram

```python
plt.figure(figsize=(10,5))
a['sepal_length'].hist(bins=20,grid=False,rwidth=0.9,edgecolor='Black')
plt.title("Histogram for Sepal Length")
plt.xlabel("Sepal Length in cm")
plt.ylabel("Frequency")
plt.show()
```

##### Interpretation of the graph :
    The histogram shows the distribution of sepal length of flowers. 
    Here in the graph most of the dataset are between 5.0 cm and 6.5 cm. 
    The graph looks symmetrical, so the data is evenly spread wothout any skewness. 
    The minimum sepal length is around 7.5 cm & the maximum sepal length is around 5.0 cm and 6.4 cm also.

#### b) Boxplot

```python
sns.boxplot(x=a['species'],y=a['sepal_length'],data=a)
plt.title("Box plot for Sepal Length")
plt.ylabel("Sepal Length in cm ")
plt.show()
```

##### Interpretation :
    The box plot shows the distribution of sepal length across different species.
    As we can see Setosa has shortest sepal length while Virginica has the longest sepal length.
    There are some outliers present in the Virginica Species.
    As we move from Setosa -> Versicolor -> Virginica, The sepal length increases.

#### c) Scatter Plot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=a)
plt.title('Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(False)
plt.show()
```

##### Interpretation :
    This Scatter Plot shows the relationship between sepal length and sepal width for different species.
    Each colour represents a different species : Blue -> Setosa , Orange -> Versicolor , Green -> Virginica.
    So as sepal length increases, we can easily identify the species.

### 7. Correlation 

```python
print(a.corr(numeric_only=True))
```

```python
cm = a.corr(numeric_only=True)
plt.figure(figsize=(10,5))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='.2f')
plt.title("Heatmap showing the correlation of numerical features")
plt.show()
```

##### Interpretation :
    This Heatmap shows the correlation between the numerical features.
    Darker shades = Positive Correlation, Lighter shades = Negative Correlation.
    We can say that with the increase in petal width, petal length also increases.
    And with increase in petal length, sepal width decreases.
