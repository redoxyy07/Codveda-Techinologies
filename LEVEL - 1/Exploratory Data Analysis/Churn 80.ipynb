### 5. Summary Statistics 

```python
num = a.select_dtypes(include=['number'])
```

#### a) Mean :
    Mean is the sum of all values divided by the number of values.

```python
print("Mean : ")
num.mean().round(2)
```

#### b) Median :
    Median is the middle most value present in a sorted sequence of values.

```python
num.median().round(2)
```

#### c) Mode :
    Mode is the value that occurs the most frequent time in a sequence of values.

```python
num.mode().iloc[0].round(2)
```

#### d) Standard Deviation
    Standard Deviation measures the amount of variation in the mean.

```python
num.std().round(2)
```

### 6. Visualizing Data Distribution 

#### a) Histogram

```python
plt.figure(figsize=(10,6))
sns.histplot(x=a['Customer service calls'], bins=20, edgecolor='Black')
plt.title("Histogram for Customer Service Calls")
plt.xlabel("Number of customer service calls")
plt.ylabel("Frequency")
plt.show()
```

#### Interpretation :
    The Histogram shows the distribution of Customer Service Calls.
    Here we can see that mostly 0-3 number of times Customer Service Calls are made and very few users have called arounf 7-9 times.
    The maximum number of service calls made is 1 times and very rare calls are made upto 7 or more times.
    The values in the graph is mostly towards the left side that indicates that the graph is 'Right Skewed' which means generally less number of calls are made by the customers.

#### b) Box Plot

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=a['Total day minutes'], x=a['Churn'])
plt.title("Box plot for Total day minutes by churn")
plt.ylabel("Total day minutes")
plt.xlabel("Churn status")
plt.show()
```

#### Interpretation :
    The box plot shows the distribution of Total day minutes by Churn Status.
    The box plot for Churned customers has higher spread than the Unchurned customers.
    Through this graph, it is observed that customers with more total day minutes are most likely to Churned.
    Though there are some outliers present with less day minutes by Unchurned customers.
    

#### c) Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=a['Total day minutes'], y=a['Total day charge'], hue=a['Churn'])
plt.title("Scatter plot for Total day minutes vs Total day charge")
plt.xlabel("Total Day Minutes")
plt.ylabel("Total Day Charge")
plt.show()
```

#### Interpretation :
    The scatter plot shows the relationship between Total Day Minutes vs Total Day Charge by Churn status.
    This graph indicates that there's a strong correlation between these features, as Total Day Minutes increases, Total Day Charge also increases.
    Here the Blue color indicates the Unchurned Customer means didn't leave the company and the orange indicates Churned Customers.
    "So, it is found that the customer with higher number of Total Day Minutes and Total Day Charge are likely to Churn."

### 7. Correlation between numerical features

```python
print(a.corr(numeric_only=True))
```

```python
cm = a.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='.2f')
plt.title("Heatmap for correlation between numeric features")
plt.show()
```

#### Interpretation :
    Heatmap is used to show the correlation between different numeric features.
    From the heatmap it is seen that Churn has very less correlation with other features.
    And features like :
        1. Total Day Charge - Total Day Minutes
        2. Total Eve Charge - Total Eve Minutes
        3. Total Night Charge - Total Night Minutes
        4. Total Intl Charge - Total Intl Minutes    
    are highly correlated to each others.
