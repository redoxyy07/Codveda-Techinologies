```python
a.dtypes
```

### 5. Summary Statistics

```python
a_num = a.select_dtypes(include=['number'])
```

#### a) Mean :
    Mean is the sum of all values divided by the total numbers of values.

```python
print("Mean of columns :")
a_num.mean().round(2)
```

#### b) Median :
    Median is the middle most value of a sorted sequence of values.

```python
print("Median : " )
a_num.median()
```

#### c) Mode :
    Mode is the value that occurs most of the times in a sequence of values.

```python
print("Mode : ")
a_num.mode().iloc[0]
```

#### d) Standard Deviation :
    Standard Deviation is the amount of variation from the mean.

```python
print("Standard Deviation : ")
a_num.std().round(2)
```

### 6. Visualizing Data Distribution

#### a) Histogram

```python
plt.figure(figsize=(8,5))
a['close'].hist(bins=20, grid=False, rwidth=0.9, edgecolor='Black')
plt.title("Histogram showing Closing Prices")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.show
```

##### Interpretation :
    This Histogram represents the Distribution of Closing Price of stocks.
    The most closing price is around 0 - 150.
    The highest frequency of Closing price of stocks is around 90-100.
    As we can clearly see that most of the datas are in the left side, which indicate that the graph is Right Skewed.

#### b) Box Plot :

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=a['close'])
plt.title("Box Plot showing the Closing Price")
plt.ylabel("Closing Price")
plt.show()
```

##### Interpretation :
    The Box Plot shows the distribution of Closing Price of stocks.
    This Box plot clearly shows that there any many number of outliers present in the dataset.
    And the outliers have higher closing prices than others.

#### c) Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=a['open'], y=a['close'])
plt.title("Scatter Plot of Opening Price vs Closing Price")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()          
```

##### Interpretation :
    This Scatter Plot graph shows the distribution of Open Price of stocks vs Closing Price of stocks.
    The graph is found as like a Linear Graph.
    Here we observe that with increasing of closing price, opening price also increases.
    There are few outliers but overall, the data aligns a straight line.

### 7. Correlation :

```python
print(a.corr(numeric_only=True))
```

```python
plt.figure(figsize=(12,6))
cm = a.corr(numeric_only=True)
sns.heatmap(cm, annot=True, cmap='Blues', fmt='.2f')
plt.title("Heatmap for numerical features")
plt.show()
```

##### Interpretation :
    This Heatmap shows the correlation between numerical features.
    Darker shades = Strong Correlation, Lighter shades = Weak Correlation.
    Through this graph we observe that 'Opening Price', 'High Price', 'Low Price', 'Closing price' of stocks are highly correlated to each other.
    But Volume has a weaker correlation with the other features.
