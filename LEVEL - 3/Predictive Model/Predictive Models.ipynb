### 1. Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

#### c. Encoding Categorical Features

```python
le = LabelEncoder()
a['Churn'] = le.fit_transform(a['Churn'])
```

### 4. Train Test Split

```python
x1 = a.drop(columns='Churn')
y1 = a['Churn']
```

```python
x1.head()
```

```python
y1.head()
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.2, random_state=42)
```

## 5. Model Implementation, Hyperparameter Tuning & Evaluation

## 

## Decision Tree

```python
dec = DecisionTreeClassifier()
dec.fit(x_train1,y_train1)
```

```python
y_pred1 = dec.predict(x_test1)
```

#### Evaluation :

```python
print("Decision Tree Evaluation : \n")
print("Accuracy : ", accuracy_score(y_test1, y_pred1))
print("Precision : ", precision_score(y_test1, y_pred1))
print("Recall : ", recall_score(y_test1, y_pred1))
print("F1 - Score : ", f1_score(y_test1,y_pred1))
```

#### HyperParameter Tuning :

```python
param_grid_dec = {
    'splitter' : ['best','random'],
    'max_depth' : [10,20,None],
    'min_samples_split' : [2,3,4,5],
    'min_samples_leaf' : [1,2,3]
    }

grid_dec = GridSearchCV( DecisionTreeClassifier(), param_grid_dec, cv=3, scoring='accuracy')
grid_dec.fit(x_train1, y_train1)
```

```python
print("Best Parameters for Decision Trees : ",grid_dec.best_params_)
```

#### Evaluation based on HyperParameter Tuning

```python
best_dec = grid_dec.best_estimator_
```

```python
y_pred1_best = best_dec.predict(x_test1)
```

```python
print("Decision Tree Evaluation using Hyperparameter Tuning : \n")
print("Accuracy : ", accuracy_score(y_test1, y_pred1_best))
print("Precision : ", precision_score(y_test1, y_pred1_best))
print("Recall : ", recall_score(y_test1, y_pred1_best))
print("F1 - Score : ", f1_score(y_test1,y_pred1_best))
```

## Logistic Regression

```python
x2 = a.drop(columns='Churn')
y2 = a['Churn']
```

```python
x_train2, x_test2, y_train2, y_test2 = train_test_split(x1, y1, test_size=0.2, random_state=42)
```

```python
print("Best parameters for Logistic Regression : \n",grid_lr.best_params_)
```

```python
y_pred2_best = best_lr.predict(x_test2)
```

#### Evaluation based on Hyperparameter Tuning

### Random Forest

```python
x3 = a.drop(columns='Churn')
y3 = a['Churn']
```

```python
x_train3, x_test3, y_train3, y_test3 = train_test_split(x3, y3, test_size=0.2, random_state=42)
```
