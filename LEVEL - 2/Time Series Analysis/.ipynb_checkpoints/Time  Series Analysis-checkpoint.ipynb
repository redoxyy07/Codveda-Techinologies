# Task 2 : Time Series Analysis

## Objectives :
    - Plot time-series data and identify patterns.
    - Decompose the series into trends, seasonality, and residuals using statsmodel.
    - Perform moving average smoothing and plot the results.

### 1. Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

### 2. Reading the CSV file

```python
a = pd.read_csv(r"C:\Users\Rudra Pratap Swain\OneDrive\Desktop\CodVeda\DATASETS\2) Stock Prices Data Set.csv")
print(a)
```

```python
a['date'] = pd.to_datetime(a['date'])
a.set_index('date', inplace=True)

a.head()
```

### 3. Filter data for a Single Stock Symbol

```python
symbol = a['symbol'].unique()
print("Available symbols : ",symbol)
```

```python
selected_symbol = symbol[0]
```

```python
symbol_data = a[a['symbol'] == selected_symbol]['close']
```

### 4. Plot Time Series Data

```python
plt.figure(figsize=(12,6))
plt.plot(symbol_data)
plt.title(f'Time Series Plot - Close Price for Symbol {selected_symbol}')
plt.xlabel("Date")
plt.ylabel("Standardize Close Price")
plt.show()
```

### 5. Decompose the Time Series

```python
res = seasonal_decompose(symbol_data, model='additive', period=30)
res.plot()
plt.tight_layout()
plt.show()
```

### 6. Moving Average Smoothing

```python
mov_avg = symbol_data.rolling(window=30).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(symbol_data, label='Original')
plt.plot(mov_avg, label='30-Day Moving Average', color='red')
plt.title(f'Moving Average Smoothing - Close Price for Symbol {selected_symbol}')
plt.xlabel("Date")
plt.ylabel("Standardized Close Price")
plt.legend()
plt.show()
```

### Interpretation :

##### Plotted the Time Series of Stock Close Price - Observed fluctuations, growth phases, and sharp declines.
##### Decomposed the series into :
        1. Trend : Captured the general growth or decline over time.
        2. Seasonality : Periodic patterns that may repeat over months or years.
        3. Residuals : Irregular, random noise remaining after removing trend & seasonality.
##### Appelied Moving Average Smoothing - reduced noise, clarified the long-term trend, and make the overall price behavior clearer.
